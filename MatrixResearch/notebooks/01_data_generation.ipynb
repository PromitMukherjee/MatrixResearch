```python
import os
import sys
sys.path.insert(0, '/content/MatrixResearch/MatrixResearch')
from src.datasets.matrix_dataset import DynamicMatrixDataset
import matplotlib.pyplot as plt

dataset = DynamicMatrixDataset(num_samples=5, size=16, matrix_type='random')
sample = dataset[0]
A, B, C = sample['A'], sample['B'], sample['C']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(A.numpy(), cmap='viridis')
axes[0].set_title('Matrix A')
axes[1].imshow(B.numpy(), cmap='viridis')
axes[1].set_title('Matrix B')
axes[2].imshow(C.numpy(), cmap='viridis')
axes[2].set_title('Matrix C = A @ B')
plt.show()
```
