```python
from src.datasets.matrix_dataset import DynamicMatrixDataset
import torch

# Create 1000 fixed matrices
fixed_data = []

dataset = DynamicMatrixDataset(
    num_samples=1000,
    size=16,
    matrix_type="low_rank"
)

for i in range(1000):
    sample = dataset[i]
    fixed_data.append(sample["A"])

print("Fixed dataset created:", len(fixed_data))
```

```python
from src.models.autoencoder import MatrixAutoencoder
import torch.nn as nn

device = "cuda"

model = MatrixAutoencoder(
    size=16,
    latent_dim=128,
    hidden_dims=[1024,512,256]
).to(device)

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)

criterion = nn.MSELoss()
```

```python
for epoch in range(20):

    total_loss = 0

    model.train()

    for A in fixed_data:

        A = A.unsqueeze(0).to(device)

        optimizer.zero_grad()

        A_rec = model(A)

        loss = criterion(A_rec, A)

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(fixed_data)

    print(
        f"Epoch {epoch+1:02d} "
        f"Loss = {avg_loss:.6f}"
    )
```

```python
A = fixed_data[0].unsqueeze(0).to(device)

with torch.no_grad():

    z = model.encode(A)

    A_rec = model(A)

print("Input Std :", A.std().item())
print("Output Std:", A_rec.std().item())

print("Latent Std:", z.std().item())
```
